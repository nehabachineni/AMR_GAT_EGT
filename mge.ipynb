```python
import os
import subprocess
import pandas as pd
import pandas.errors

input_dir = "/content/drive/MyDrive/poc/all_fastas_here"
script_path = "/content/drive/MyDrive/poc/mobileOG-db/mobileOG-pl/mobileOGs-pl-kyanite.sh"
diamond_db = "/content/drive/MyDrive/poc/mge_files/mobileOG-db-beatrix-1.6.dmnd"
metadata_file = "/content/drive/MyDrive/poc/mobileOG-db/metadata.tsv"

fasta_files = sorted([f for f in os.listdir(input_dir) if f.endswith(".fna")])

for fasta_file in fasta_files:
    fasta_path = os.path.join(input_dir, fasta_file)
    print(f"\n--- Running on: {fasta_file}")

    # 1. Check if .fna is empty
    if os.stat(fasta_path).st_size == 0:
        print("   Skipping empty .fna file")
        continue

    # 2. Check if .fna starts with '>'
    with open(fasta_path, 'r') as f:
        first_line = f.readline().strip()
        if not first_line.startswith(">"):
            print("   File format error: First line is not a FASTA header (>)")
            continue

    log_out = os.path.join(input_dir, fasta_file + ".log")
    log_err = os.path.join(input_dir, fasta_file + ".err")

    try:
        with open(log_out, "w") as out_file, open(log_err, "w") as err_file:
            subprocess.run(
                [
                    "bash", script_path,
                    "-i", fasta_path,
                    "-k", "1",
                    "-e", "1e-5",
                    "-p", "30",
                    "-d", diamond_db,
                    "-q", "60",
                    "-m", metadata_file
                ],
                check=True,
                stdout=out_file,
                stderr=err_file
            )

        # 3. Check if .faa file exists and starts properly
        faa_path = fasta_path + ".faa"
        if os.path.exists(faa_path):
            with open(faa_path, 'r') as faa:
                line = faa.readline().strip()
                if line == "" or not line.startswith(">"):
                    print("   Skipping: .faa file is empty or invalid (bad header)")
                    continue
        else:
            print("   No .faa file created by script")
            continue

        # 4. Process the .tsv output
        tsv_file = fasta_path + ".tsv"
        if os.path.exists(tsv_file):
            if os.stat(tsv_file).st_size == 0:
                print("    .tsv exists but is empty")
            else:
                try:
                    df = pd.read_csv(tsv_file, sep="\t", header=None)
                    print(f"   Found {len(df)} MGEs")
                except pd.errors.EmptyDataError:
                    print("    .tsv exists but unreadable")
        else:
            print("   No .tsv output found")

    except subprocess.CalledProcessError:
        print("   Error while running the script (subprocess failed)")
        print(f"     Check logs:\n     ├── stdout: {log_out}\n     └── stderr: {log_err}")
```

```python
import os
import shutil
import re
from collections import defaultdict

def get_prefix(folder_name):
    match = re.match(r"^(\d{3})", folder_name)
    return match.group(1) if match else None

def is_complete(folder_path):
    return len([
        f for f in os.listdir(folder_path)
        if os.path.isfile(os.path.join(folder_path, f))
    ]) == 6

def process_folders_ascending(base_path):
    folders = [f for f in os.listdir(base_path)
               if os.path.isdir(os.path.join(base_path, f)) and get_prefix(f)]

    # Sort folders based on their numerical prefix
    folders.sort(key=lambda x: int(get_prefix(x)))

    prefix_groups = defaultdict(list)

    for folder in folders:
        prefix = get_prefix(folder)
        if prefix:
            prefix_groups[prefix].append(folder)

    processed_prefixes = set()

    for prefix_int in range(1, 726):  # Loop from 001 to 725
        prefix = f"{prefix_int:03d}"  # Format as three-digit string (e.g., "001")

        if prefix in prefix_groups and prefix not in processed_prefixes:
            group = prefix_groups[prefix]
            merged_folder = os.path.join(base_path, prefix)
            merged = False

            if len(group) > 1:
                # Merge into one main folder named just the prefix if not already exists
                if not os.path.exists(merged_folder):
                    os.makedirs(merged_folder)
                    merged = True

                for folder in group:
                    folder_path = os.path.join(base_path, folder)
                    for file in os.listdir(folder_path):
                        file_path = os.path.join(folder_path, file)
                        if os.path.isfile(file_path):
                            target_path = os.path.join(merged_folder, file)
                            if not os.path.exists(target_path):
                                shutil.copy(file_path, target_path)
                                if merged:
                                    print(f"{prefix}: Merging '{folder}' -> '{prefix}'")
                if merged and is_complete(merged_folder):
                    new_folder_name = f"{prefix}_done"
                    new_folder_path = os.path.join(base_path, new_folder_name)
                    try:
                        os.rename(merged_folder, new_folder_path)
                        print(f"{prefix}: Merged & Renamed '{prefix}' -> '{new_folder_name}'")
                    except OSError as e:
                        print(f"{prefix}: Error renaming merged '{prefix}': {e}")
                elif merged:
                    print(f"{prefix}: Merged but incomplete (<6 files)")

            elif len(group) == 1:
                # Only one folder — check if it's complete and rename
                single_folder_path = os.path.join(base_path, group[0])
                if is_complete(single_folder_path):
                    new_folder_name = f"{prefix}_done"
                    new_folder_path = os.path.join(base_path, new_folder_name)
                    try:
                        os.rename(single_folder_path, new_folder_path)
                        print(f"{prefix}: Single folder, Renamed '{group[0]}' -> '{new_folder_name}'")
                    except OSError as e:
                        print(f"{prefix}: Error renaming single '{group[0]}': {e}")
                else:
                    print(f"{prefix}: Single folder, incomplete (<6 files)")

            processed_prefixes.add(prefix)

        elif prefix not in prefix_groups:
            print(f"{prefix}: -- No folder found with this prefix")

# Run the integrated function
process_folders_ascending("/content/drive/MyDrive/poc/MGE Results 725")
```

```python
import os
import time
from datetime import datetime, timedelta
from google.colab import drive
import pytz
import re

# Mount Google Drive
drive.mount('/content/drive')

def get_prefix(folder_name):
    match = re.match(r"^(\d{3})", folder_name)
    return match.group(1) if match else None

def has_done_suffix(folder_name):
    return folder_name.endswith("_done")

def process_modified_folders_ist(base_path, target_time_str_ist):
    """
    Checks folders modified after the specified IST time and adds the '_done' suffix if it's not already present.

    Args:
        base_path (str): The path to the directory to search.
        target_time_str_ist (str): A string representing the target time in IST in HH:MM AM/PM format (e.g., "07:51 PM").
    """
    ist_timezone = pytz.timezone('Asia/Kolkata')
    utc_timezone = pytz.utc

    # Convert the target IST time to UTC
    target_time_naive = datetime.now().replace(hour=int(target_time_str_ist[:2]),
                                                minute=int(target_time_str_ist[3:5]),
                                                second=0,
                                                microsecond=0)
    if "PM" in target_time_str_ist and target_time_naive.hour < 12:
        target_time_naive += timedelta(hours=12)
    elif "AM" in target_time_str_ist and target_time_naive.hour == 12:
        target_time_naive -= timedelta(hours=12)
    target_datetime_ist = ist_timezone.localize(target_time_naive.replace(tzinfo=None))
    target_datetime_utc = target_datetime_ist.astimezone(utc_timezone)

    print(f"Target datetime for modification check (UTC): {target_datetime_utc}")

    for item in os.listdir(base_path):
        item_path = os.path.join(base_path, item)
        if os.path.isdir(item_path) and get_prefix(item):
            modified_timestamp = os.path.getmtime(item_path)
            modified_datetime_utc = datetime.fromtimestamp(modified_timestamp, utc_timezone)
            print(f"Checking folder: {item}, Modified on (UTC): {modified_datetime_utc}")

            if modified_datetime_utc > target_datetime_utc and not has_done_suffix(item):
                new_folder_name = f"{item}_done"
                new_folder_path = os.path.join(base_path, new_folder_name)
                try:
                    os.rename(item_path, new_folder_path)
                    print(f"  - Modified after {target_time_str_ist} IST (in UTC), Renamed '{item}' to '{new_folder_name}'")
                except OSError as e:
                    print(f"  - Error renaming '{item}': {e}")
            elif has_done_suffix(item):
                print(f"  - Folder '{item}' already has the '_done' suffix.")
            else:
                print(f"  - Folder '{item}' not modified after {target_time_str_ist} IST (in UTC).")

if __name__ == "__main__":
    base_directory = "/content/drive/MyDrive/poc/MGE Results 725"  # Replace with your actual base path
    target_time_ist = "07:51 PM"

    process_modified_folders_ist(base_directory, target_time_ist)
    print("\nProcessing complete.")
```

```python
# prompt: for all folders with suffix _done in "/content/drive/MyDrive/poc/MGE Results 725"  and list them numerically ascending  order , count number of files in each folder  and check if they arent empty , order it ascendingly , 001_done first ,,,, then 002_done ,,,, etc , print if any file in any folder is empty , if not print 'all 6 good' for each folder

def check_folders_and_files(directory):
    """
    Checks folders with the suffix "_done", lists them numerically ascending,
    counts files in each, and verifies they aren't empty.
    """

    done_folders = []
    for filename in os.listdir(directory):
        if os.path.isdir(os.path.join(directory, filename)) and filename.endswith("_done"):
            match = re.match(r"^(\d{3})_done$", filename)
            if match:
                done_folders.append(int(match.group(1)))

    done_folders.sort()

    for folder_num in done_folders:
        folder_name = f"{folder_num:03d}_done"
        folder_path = os.path.join(directory, folder_name)
        files_in_folder = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
        num_files = len(files_in_folder)

        print(f"Folder: {folder_name}, File Count: {num_files}")

        empty_files_found = False
        if num_files == 6:
          for file in files_in_folder:
              file_path = os.path.join(folder_path, file)
              if os.stat(file_path).st_size == 0:
                  print(f"  - Empty file found: {file}")
                  empty_files_found = True
          if not empty_files_found:
              print("  - all 6 good")
        else:
            print(f"  - Error: Expected 6 files, found {num_files}")

# Example usage
check_folders_and_files("/content/drive/MyDrive/poc/MGE Results 725")
```
